```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, multivariate_normal
from scipy.signal import convolve2d
from scipy.interpolate import RegularGridInterpolator
from pathlib import Path
```

```python
color_max = "#D81B60"
color_60 = "#1E88E5"
color_50 = "#FFC107"
```

```python
# https://stats.stackexchange.com/questions/64680/how-to-determine-quantiles-isolines-of-a-multivariate-normal-distribution
q2sigma = lambda q: - 2 * np.log(1 - q)
```

```python
d_board = 451e-3 # (m)
d_bull = 31.8e-3 # (m)
d_dbull = 12.7e-3 # (m)
d_triple = 214e-3 # (m)
d_double = 340e-3 # (m)
w_ring = 8e-3 # (m)
# w_wire = 1.5e-3 # (m)
alpha = 18 # (deg)

dx = 1e-3 # (m)
dy = 1e-3 # (m)
x = np.arange(- d_board / 2, d_board / 2, dx)
y = np.arange(- d_board / 2, d_board / 2, dy)

x_g, y_g = np.meshgrid(x, y, indexing="xy")

r_g = np.sqrt(x_g ** 2 + y_g ** 2)
phi_g = 180 / np.pi * np.arctan2(y_g, x_g)
phi_g[phi_g < 0] += 360
```

```python
factors = np.ones_like(r_g)
points = 0 * np.ones_like(phi_g)

points[(r_g <= d_dbull / 2)] = 50
points[(r_g > d_dbull / 2) & (r_g <= d_bull / 2)] = 25

points[(r_g > d_bull / 2) & ((phi_g > 360 - alpha / 2) | (phi_g <= alpha / 2))] = 6
points[(r_g > d_bull / 2) & (phi_g > alpha / 2) & (phi_g <= 3 * alpha / 2)] = 13
points[(r_g > d_bull / 2) & (phi_g > 3 * alpha / 2) & (phi_g <= 5 * alpha / 2)] = 4
points[(r_g > d_bull / 2) & (phi_g > 5 * alpha / 2) & (phi_g <= 7 * alpha / 2)] = 18
points[(r_g > d_bull / 2) & (phi_g > 7 * alpha / 2) & (phi_g <= 9 * alpha / 2)] = 1
points[(r_g > d_bull / 2) & (phi_g > 9 * alpha / 2) & (phi_g <= 11 * alpha / 2)] = 20
points[(r_g > d_bull / 2) & (phi_g > 11 * alpha / 2) & (phi_g <= 13 * alpha / 2)] = 5
points[(r_g > d_bull / 2) & (phi_g > 13 * alpha / 2) & (phi_g <= 15 * alpha / 2)] = 12
points[(r_g > d_bull / 2) & (phi_g > 15 * alpha / 2) & (phi_g <= 17 * alpha / 2)] = 9
points[(r_g > d_bull / 2) & (phi_g > 17 * alpha / 2) & (phi_g <= 19 * alpha / 2)] = 14
points[(r_g > d_bull / 2) & (phi_g > 19 * alpha / 2) & (phi_g <= 21 * alpha / 2)] = 11
points[(r_g > d_bull / 2) & (phi_g > 21 * alpha / 2) & (phi_g <= 23 * alpha / 2)] = 8
points[(r_g > d_bull / 2) & (phi_g > 23 * alpha / 2) & (phi_g <= 25 * alpha / 2)] = 16
points[(r_g > d_bull / 2) & (phi_g > 25 * alpha / 2) & (phi_g <= 27 * alpha / 2)] = 7
points[(r_g > d_bull / 2) & (phi_g > 27 * alpha / 2) & (phi_g <= 29 * alpha / 2)] = 19
points[(r_g > d_bull / 2) & (phi_g > 29 * alpha / 2) & (phi_g <= 31 * alpha / 2)] = 3
points[(r_g > d_bull / 2) & (phi_g > 31 * alpha / 2) & (phi_g <= 33 * alpha / 2)] = 17
points[(r_g > d_bull / 2) & (phi_g > 33 * alpha / 2) & (phi_g <= 35 * alpha / 2)] = 2
points[(r_g > d_bull / 2) & (phi_g > 35 * alpha / 2) & (phi_g <= 37 * alpha / 2)] = 15
points[(r_g > d_bull / 2) & (phi_g > 37 * alpha / 2) & (phi_g <= 39 * alpha / 2)] = 10

factors[r_g > d_double / 2] = 0
factors[(r_g <= d_double / 2) & (r_g > d_double / 2 - w_ring)] = 2
factors[(r_g <= d_triple / 2) & (r_g > d_triple / 2 - w_ring)] = 3

points *= factors
```

```python
plt.figure(figsize=(4, 4))

plt.pcolormesh(x_g, y_g, points, rasterized="True", cmap="magma")

points_order = (6, 13, 4, 18, 1, 20, 5, 12, 9, 14, 11, 8, 16, 7, 19, 3, 17, 2, 15, 10)
for ii in range(len(points_order)):
    plt.text((d_double / 2 + (d_board - d_double) / 4) * np.cos(ii * np.pi / 180 * alpha), (d_double / 2 + (d_board - d_double) / 4) * np.sin(ii * np.pi / 180 * alpha), 
             points_order[ii], 
             horizontalalignment="center",
             verticalalignment="center",
             fontsize=12, 
             color=(1, 1, 1))
    
plt.gca().axis("off")
plt.gca().set_aspect(1)

plt.savefig(Path("..", "content", "img", "dartboard_points.svg"), dpi=600)
```

```python
mu = 0 # (m)
sigma = 1e-3 * np.linspace(10, 100, 46) # (m)
sigma = np.append(sigma, 200e-3) # (m)
n_sigma = 3

points_expected = np.zeros((*np.shape(x_g), len(sigma)))

r_60 = d_triple / 2 - w_ring / 2
phi_60 = 90
x_60 = r_60 * np.cos(np.pi * phi_60 / 180)
y_60 = r_60 * np.sin(np.pi * phi_60 / 180)
points_expected_60 = np.ones_like(sigma)

r_50 = 0
phi_50 = 0
x_50 = r_50 * np.cos(np.pi * phi_50 / 180)
y_50 = r_50 * np.sin(np.pi * phi_50 / 180)
points_expected_50 = np.ones_like(sigma)

x_max = np.ones_like(sigma)
y_max = np.ones_like(sigma)
points_expected_max = np.ones_like(sigma)

for ii in range(len(sigma)):
    x_psf = np.arange(-n_sigma * sigma[ii], n_sigma * sigma[ii], dx)
    y_psf = np.arange(-n_sigma * sigma[ii], n_sigma * sigma[ii], dy)
    x_psf, y_psf = np.meshgrid(x_psf, y_psf, indexing="xy")
    r_psf = np.sqrt(x_psf ** 2 + y_psf ** 2)
    
    psf = norm(loc=mu, scale=sigma[ii]).pdf(r_psf)
    psf /= np.sum(psf)

    points_expected[:, :, ii] = convolve2d(points, psf, mode="same", boundary="fill", fillvalue=0)
    
    points_expected_60[ii] = RegularGridInterpolator((x, y), points_expected[:, :, ii].T)((x_60, y_60))
    points_expected_50[ii] = RegularGridInterpolator((x, y), points_expected[:, :, ii].T)((x_50, y_50))
    
    idx_max = np.unravel_index(np.argmax(points_expected[:, :, ii]), np.shape(points_expected[:, :, ii]))    
    points_expected_max[ii] = points_expected[:, :, ii][idx_max]
    x_max[ii] = x_g[idx_max]
    y_max[ii] = y_g[idx_max]
```

```python
for ii in range(len(sigma)):    
    if np.isin(sigma[ii], [10e-3, 20e-3, 50e-3, 100e-3, 200e-3]):
        plt.figure(figsize=(4, 4))
                  
        plt.text(0.05, 0.95, f"$\\sigma\,=\,{sigma[ii] * 1e3:.0f}\,$mm", 
                 horizontalalignment="left",
                 verticalalignment="top",
                 color=(1, 1, 1),
                 fontsize=12,
                 transform=plt.gca().transAxes)
        
        plt.text(0.95, 0.95, f"$\\langle p_\\mathrm{{max}}\\rangle ={points_expected_max[ii]:.1f}\,$pts", 
                 horizontalalignment="right",
                 verticalalignment="top",
                 color=color_max,
                 fontsize=12,
                 transform=plt.gca().transAxes)
        plt.text(0.95, 0.875, f"$\\langle p_{{60}}\\rangle ={points_expected_60[ii]:.1f}\,$pts", 
                 horizontalalignment="right",
                 verticalalignment="top",
                 color=color_60,
                 fontsize=12,
                 transform=plt.gca().transAxes)
        plt.text(0.95, 0.80, f"$\\langle p_{{50}}\\rangle ={points_expected_50[ii]:.1f}\,$pts", 
                 horizontalalignment="right",
                 verticalalignment="top",
                 color=color_50,
                 fontsize=12,
                 transform=plt.gca().transAxes)        

        plt.pcolormesh(x_g, y_g, points_expected[:, :, ii], rasterized="True", cmap="magma")
        plt.scatter(x_60, y_60, edgecolor=(0, 0, 0), facecolor=color_60, s=50, marker="X", zorder=15)
        plt.scatter(x_50, y_50, edgecolor=(0, 0, 0), facecolor=color_50, s=50, marker="X", zorder=15)
        plt.scatter(x_max[ii], y_max[ii], edgecolor=(0, 0, 0), facecolor=color_max, s=50, marker="X", zorder=15)
        
        plt.gca().axis("off")
        plt.gca().set_aspect(1)
        
        plt.savefig(Path("..", "content", "img", f"xpoints_sigma{sigma[ii] * 1e3:.0f}mm.svg", dpi=600, bbox_inches="tight"))
```

```python
for ii in range(len(sigma)):
    if np.isin(sigma[ii], [10e-3, 20e-3, 50e-3, 100e-3, 200e-3]):
        plt.figure(figsize=(4, 4))
                  
        plt.text(0.05, 0.95, f"$\\sigma\,=\,{sigma[ii] * 1e3:.0f}\,$mm", 
                 horizontalalignment="left",
                 verticalalignment="top",
                 color=(1, 1, 1),
                 fontsize=12,
                 transform=plt.gca().transAxes)
        
        plt.text(0.95, 0.95, f"$\\langle p_\\mathrm{{max}}\\rangle ={points_expected_max[ii]:.1f}\,$pts", 
                 horizontalalignment="right",
                 verticalalignment="top",
                 color=color_max,
                 fontsize=12,
                 transform=plt.gca().transAxes)
        plt.text(0.95, 0.875, f"$\\langle p_{{60}}\\rangle ={points_expected_60[ii]:.1f}\,$pts", 
                 horizontalalignment="right",
                 verticalalignment="top",
                 color=color_60,
                 fontsize=12,
                 transform=plt.gca().transAxes)
        plt.text(0.95, 0.8, f"$\\langle p_{{50}}\\rangle ={points_expected_50[ii]:.1f}\,$pts", 
                 horizontalalignment="right",
                 verticalalignment="top",
                 color=color_50,
                 fontsize=12,
                 transform=plt.gca().transAxes)
        
        plt.pcolormesh(x_g, y_g, points, rasterized="True", cmap="magma")
        
        circle_1 = plt.Circle((x_max[ii], y_max[ii]), sigma[ii], facecolor="None", edgecolor=color_max)
        circle_2 = plt.Circle((x_max[ii], y_max[ii]), n_sigma * sigma[ii], facecolor="None", edgecolor=color_max, linestyle="--")
        plt.gca().add_artist(circle_1)
        plt.gca().add_artist(circle_2)

        plt.scatter(x_60, y_60, edgecolor=(0, 0, 0), facecolor=color_60, s=50, marker="X")
        plt.scatter(x_50, y_50, edgecolor=(0, 0, 0), facecolor=color_50, s=50, marker="X")
        plt.scatter(x_max[ii], y_max[ii], edgecolor=(0, 0, 0), facecolor=color_max, s=50, marker="X")
        
        plt.gca().axis("off")
        plt.gca().set_aspect(1)
        
        plt.savefig(Path("..", "content", "img", f"points_sigma{sigma[ii] * 1e3:.0f}mm.svg", dpi=600, bbox_inches="tight"))
```

```python
plt.figure(figsize=(4, 4))

plt.pcolormesh(x_g, y_g, points, rasterized="True", cmap="magma", zorder=5)

for ii in range(len(sigma)):
    if np.isin(sigma[ii], [10e-3, 20e-3, 50e-3, 100e-3, 200e-3]):
        plt.scatter(x_max[ii], y_max[ii], edgecolor=(0, 0, 0), facecolor=color_max, s=50, marker="X", zorder=15)
        
        if sigma[ii] == sigma[-1]:
            h_align = "left"
        else:
            h_align = "right"
        plt.text(x_max[ii], y_max[ii], f"$\\sigma\,=\,{sigma[ii] * 1e3:.0f}\,$mm",
            horizontalalignment=h_align,
            verticalalignment="center",
            color=(1, 1, 1),
            fontsize=10,
            zorder=20
            )    
    else:
        plt.scatter(x_max[ii], y_max[ii], color=(0.6, 0.6, 0.6), s=10, zorder=10)
        
plt.gca().axis("off")
plt.gca().set_aspect(1)

plt.savefig(Path("..", "content", "img", "points_sigmas.svg", dpi=600, bbox_inches="tight"))
```

```python
plt.figure(figsize=(6, 4))

plt.plot(1e3 * sigma, 100 * (points_expected_max - points_expected_60) / points_expected_60, 
         color=color_60, 
         label="$\\Delta p = \\langle p_\\mathrm{max}\\rangle \,/\, \\langle p_{60}\\rangle - 1$")
plt.plot(1e3 * sigma, 100 * (points_expected_max - points_expected_50) / points_expected_50, color=color_50, 
         label="$\\Delta p = \\langle p_\\mathrm{max}\\rangle \,/\, \\langle p_{50}\\rangle - 1$")

plt.axhline(5, linestyle="--", color=(0, 0, 0))

plt.xlabel("$\\sigma$ (mm)", fontsize=12)
plt.ylabel("$\\Delta p$ (%)", fontsize=12)
plt.xlim([0, 100])
plt.gca().tick_params(axis="both", which="major", labelsize=10)
plt.grid(alpha=0.2)

plt.legend()

plt.savefig(Path("..", "content", "img", f"advantage_sigma_all.svg", dpi=600, bbox_inches="tight"))
```

```python
plt.figure(figsize=(4, 4))

points_delta = points_expected_max[None, :] - points_expected_60[:, None]

plt.pcolormesh(1e3 * sigma, 1e3 * sigma, points_delta, rasterized="True", cmap="RdBu_r", vmin=-np.max(np.abs(points_delta)), vmax=np.max(np.abs(points_delta)), zorder=5)
plt.plot(1e3 * sigma, 1e3 * sigma, zorder=10, linestyle="--", color=(0, 0, 0))
contour = plt.contour(1e3 * sigma, 1e3 * sigma, points_expected_max[None, :] - points_expected_60[:, None], levels=[0], colors=[color_60], zorder=15)

vertices = contour.collections[0].get_paths()[0].vertices
sigma_60_1 = vertices[:, 0]
sigma_60_2 = vertices[:, 1]

plt.text(0.5, 0.5, "$\\sigma_1 = \\sigma_2$", color=(0, 0, 0), 
         verticalalignment="center",
         horizontalalignment="center",
         fontsize=10, 
         transform=plt.gca().transAxes, 
         zorder=20)
plt.text(0.5, 0.3, "$\\langle p_\\mathrm{max}(\\sigma_1) \\rangle = \\langle p_{60}(\\sigma_2) \\rangle$", 
         color=color_60,
         verticalalignment="center",
         horizontalalignment="center",
         fontsize=10, 
         transform=plt.gca().transAxes, 
         zorder=20)

plt.xlabel("$\\sigma_1$ (mm)", fontsize=12)
plt.ylabel("$\\sigma_2$ (mm)", fontsize=12)
plt.xlim([1e3 * sigma[0], 1e3 * sigma[-2]])
plt.ylim([1e3 * sigma[0], 1e3 * sigma[-2]])
plt.title("$\\langle p_\\mathrm{max}(\\sigma_1) \\rangle - \\langle p_{60}(\\sigma_2) \\rangle$")

plt.savefig(Path("..", "content", "img", f"advantage_max_60.svg", dpi=600, bbox_inches="tight"))
```
